```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Read teh image
img = cv.imread('a1images/brain_proton_density_slice.png', cv.IMREAD_GRAYSCALE)

# Step 2: Define the gaussian pulse for white matter
mu = 150
sigma = 20
x = np.linspace(0, 255, 256)
white = 255 * np.exp(-((x - mu)**2) / (2 * sigma**2))

# Step 3: Define the gaussian pulse for white matter
mu = 200
sigma = 15
x = np.linspace(0, 255, 256)
gray = 255 * np.exp(-((x - mu)**2) / (2 * sigma**2))

# Step 4: Ensure the output is in the valid range for image intensities [0, 255]
white = np.clip(white, 0, 255)
gray = np.clip(gray, 0, 255)

# Step 5: Apply the transformation
white_matter = white[img]
gray_matter = gray[img]

# Step 6: Display results
plt.figure(figsize=(10,4))

plt.subplot(2,3,1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(2,3,2)
plt.title("Transformation Curve")
plt.plot(np.arange(256), white, color='red')
plt.xlabel("Input Intensity")
plt.ylabel("Output Intensity")

plt.subplot(2,3,3)
plt.title("Accentuated white matter")
plt.imshow(white_matter, cmap='gray')
plt.axis('off')

plt.subplot(2,3,4)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(2,3,5)
plt.title("Transformation Curve")
plt.plot(np.arange(256), gray, color='red')
plt.xlabel("Input Intensity")
plt.ylabel("Output Intensity")

plt.subplot(2,3,6)
plt.title("Accentuated gray matter")
plt.imshow(gray_matter, cmap='gray')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Display the image
plt.figure(figsize=(5, 5))
plt.imshow(white_matter, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```
