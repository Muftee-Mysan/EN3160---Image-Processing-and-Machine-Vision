```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define piecewise transformation function
t1 = np.linspace(0, 50, num=51).astype('uint8')
t2 = np.linspace(100, 255, num=100).astype('uint8')
t3 = np.linspace(150, 255, num=105).astype('uint8')

# Step 2: Concatenate all segments to create the transformation array
t = np.concatenate((t1, t2, t3), axis=0).astype('uint8')
print(t.shape)

# Step 3: Read the image (Fig. 1b)
img = cv.imread('a1images/emma.jpg', cv.IMREAD_GRAYSCALE)

# Step 4: Apply the transformation
transformed_img = t[img]

# Step 5: Display results
plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Transformed")
plt.imshow(transformed_img, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title("Transformation Curve")
plt.plot(np.arange(256), t, color='red')
plt.xlabel("Input Intensity")
plt.ylabel("Output Intensity")

plt.tight_layout()
plt.show()
```
