```python
import cv2 as cv
import matplotlib.pyplot as plt

# Load the image
image_path = 'a1images/emma.jpg'
original_image = cv.imread(image_path)

# Check if the image was loaded correctly
if original_image is None:
    print(f"Error: Could not load image from {image_path}")
else:
    # Convert the image from BGR (OpenCV default) to RGB for correct color display in matplotlib
    original_image_rgb = cv.cvtColor(original_image, cv.COLOR_BGR2RGB)

    # Convert the original image to grayscale
    gray_image = cv.cvtColor(original_image, cv.COLOR_BGR2GRAY)

    # Display the original and grayscale images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(original_image_rgb)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Grayscale Image')
    plt.axis('off')

    plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the breakpoints from Fig. 1a
input_points = np.array([0, 50, 50, 150, 150, 255], dtype=np.uint8)
output_points = np.array([0, 50, 100, 255, 150, 255], dtype=np.uint8)

# Step 2: Create lookup table using interpolation
LUT = np.interp(np.arange(256), input_points, output_points).astype(np.uint8)

# Step 3: Read the image (Fig. 1b)
img = cv.imread('a1images/emma.jpg', cv.IMREAD_GRAYSCALE)

# Step 4: Apply the transformation
transformed_img = cv.LUT(img, LUT)

# Step 5: Display results
plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title("Transformed")
plt.imshow(transformed_img, cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Transformation Curve")
plt.plot(np.arange(256), LUT, color='red')
plt.xlabel("Input Intensity")
plt.ylabel("Output Intensity")

plt.tight_layout()
plt.show()
```
