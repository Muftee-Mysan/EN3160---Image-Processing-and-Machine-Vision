```python
# Import necessary libraries
import cv2 as cv  # OpenCV for image processing (not used here, but often paired with NumPy)
import numpy as np  # NumPy for array manipulation (images are represented as arrays)
import matplotlib.pyplot as plt  # Matplotlib for visualization

# Create a 6x8 grayscale image initialized with zeros (black)
# - np.zeros() creates an array filled with zeros
# - (6, 8) specifies 6 rows (height) and 8 columns (width)
# - dtype=np.uint8 ensures pixel values are 8-bit integers (0-255)
im = np.zeros((6, 8), dtype=np.uint8)

# Set a single pixel at row 2, column 3 to white (255)
# - Note: Python uses 0-based indexing, so (2, 3) is the 3rd row and 4th column
im[2, 3] = 255

# Create a figure and axis for plotting using Matplotlib
# - figsize=(6, 8) sets the display size in inches (width, height)
fig, ax = plt.subplots(1, 1, figsize=(6, 8))

# Display the image on the axis
# - imshow() renders the NumPy array as an image
# - cmap='gray' uses a grayscale color map (0=black, 255=white)
# - vmin=0 and vmax=255 fix the color scale range
ax.imshow(im, cmap='gray', vmin=0, vmax=255)

# Move the x-axis ticks to the top of the plot (uncommon but sometimes used in image processing)
ax.xaxis.set_ticks_position('top')

# Show the plot
plt.show()
```

```python
# Image Opening and Displaying

#Displaying using Matplotlib
import cv2 as cv
import matplotlib.pyplot as plt

im = cv.imread('batting.jpg')
fig, ax = plt.subplots(1, 1, figsize = (6,8))
ax.imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
ax.xaxis.set_ticks_position('top')
plt.show()
```

```python
#Displaying using OpenCV
import cv2 as cv
im = cv.imread('batting.jpg')
cv.namedWindow('Image', cv.WINDOW_NORMAL)  # Allows resizing
cv.resizeWindow('Image', 800, 600)  # Set initial window size (width, height)
cv.imshow('Image', im)
cv.waitKey(0)
cv.destroyAllWindows()
```
